### Q5: Fill the correct values for creating heatmap representing the correlation from the data?

```python
import matplotlib.pyplot as plt
import seaborn as sns
def correlation_heatmap(data):
    _, ax = plt.subplots(figsize=(25, 20))  # Sets the size of the figure
    colormap = sns.diverging_palette(220, 10, as_cmap=True)  # Defines the color palette
    
    # Calculate the correlation matrix
    corr_matrix = data.corr()
    
    # Create the heatmap using seaborn
    sns.heatmap(corr_matrix, cmap=colormap, annot=True, fmt=".2f", 
                xticklabels=corr_matrix.columns, 
                yticklabels=corr_matrix.columns,
                ax=ax, 
                linewidths=0.5, cbar_kws={"shrink": .5})
    
    # Additional options
    plt.title('Heatmap of Correlation Matrix')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()
    
```

```python
correlation_heatmap(data)
```

### Q7 : What is wrong with this code for scaling the data ?

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(data_x,data_y, random_state = 86)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
train_x = ss.fit_transform(train_x)
test_x = ss.fit_transform(test_x)
```

Ans : The fit method calculates the mean and standard deviation for scaling the data. When you use fit_transform on the test set, it recalculates and uses the mean and standard deviation of the test set. This is incorrect because the test set should be transformed using the parameters (mean and standard deviation) computed from the training set.

Correct Approach:

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split data into training and test sets
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, random_state=86)

# Initialize the StandardScaler
ss = StandardScaler()

# Fit the scaler on the training data and transform it
train_x = ss.fit_transform(train_x)

# Transform the test data with the same scaler
test_x = ss.transform(test_x)
